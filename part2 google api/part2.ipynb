```python
from ast import literal_eval
import urllib.parse
import string
import pandas as pd
import numpy as np
import math
import urllib.request
import json

google_df=pd.DataFrame(columns=['Description','Country','City','Highlights','Referenced_highlights','Formatted_address','Images_ids','Places_ids'])
data_file=pd.read_excel('part2 google api/output_final_spacy_standford_more_refined.xlsx')
data_file=data_file[11607:]
cnt=0
# key="AIzaSyDpQNelDhf-Kg1a7vOp4U7QNjw3UGn8DMk"
# quoted = urllib.parse.quote(key)
for index,row in data_file.iterrows():
    cnt+=1
    print(cnt)
    city=row['City'].strip()
    country=row['Country'].strip()
    places=literal_eval(row['Highlights'])
    ref_highlights=[]
    images_id=[]
    places_id=[]
    form_addr=[]
    for place in places:
        place_s=place.strip()
        if place_s!=city:
            search_str=(place_s+' '+city).replace(' ','+')
            u_str="https://maps.googleapis.com/maps/api/place/textsearch/json?query="+search_str+"&language=en&key=AIzaSyDpQNelDhf-Kg1a7vOp4U7QNjw3UGn8DMk"
            url = urllib.parse.quote(u_str, safe=string.printable)
            contents = urllib.request.urlopen(url).read()
#             print('contents:')
#             print(contents)
            my_json = contents.decode('utf8').replace("\'", "'")
#             print('myjson')
#             print(my_json)
            y = json.loads(my_json)
            if len(y['results'])>0:
                if len(y['results'])==1:
                    if 'place_id' in y['results'][0]:
#                         print(y['results'][0]['place_id'])
                        places_id.append(y['results'][0]['place_id'])
                    else:
#                         print('0000')
                        places_id.append('0000')
                    if 'photos' in y['results'][0]:
                        if 'photo_reference' in y['results'][0]['photos'][0]:
#                             print(y['results'][0]['photos'][0]['photo_reference'])
                            images_id.append(y['results'][0]['photos'][0]['photo_reference'])
                        else:
#                             print('0000')
                            images_id.append('0000')
                    else:
#                         print('0000')
                        images_id.append('0000')
                    if 'name' in y['results'][0]:
                        ref_highlights.append(y['results'][0]['name'])
                    else:
                        ref_highlights.append('0000')
                    if 'formatted_address' in y['results'][0]:
                        form_addr.append(y['results'][0]['formatted_address'])
                    else:
                        form_addr.append('0000')
#                     print('\n')
                else:
                    max_urt=0
                    main_result=y['results'][0]
                    for r in y['results']:
                        if 'user_ratings_total' in r:
                            if int(r['user_ratings_total'])>max_urt:
                                max_urt=int(r['user_ratings_total'])
                                main_result=r
                    if 'place_id' in main_result:
#                         print(main_result['place_id'])
                        places_id.append(main_result['place_id'])
                    else:
#                         print('0000')
                        places_id.append('0000')
                    if 'photos' in main_result:
                        if 'photo_reference' in main_result['photos'][0]:
#                             print(main_result['photos'][0]['photo_reference'])
                            images_id.append(main_result['photos'][0]['photo_reference'])
                        else:
#                             print('0000')
                            images_id.append('0000')
                    else:
#                         print('0000')
                        images_id.append('0000')
                    if 'name' in main_result:
                        ref_highlights.append(main_result['name'])
                    else:
                        ref_highlights.append('0000')
                    if 'formatted_address' in main_result:
                        form_addr.append(main_result['formatted_address'])
                    else:
                        form_addr.append('0000')
#                     print('\n')
    google_df.loc[len(google_df)]=[row['Description'],row['Country'],row['City'],row['Highlights'],ref_highlights,form_addr,images_id,places_id]
```
